```python
import matplotlib.pyplot as plt
plt.style.use('dark_background')

# Data for the first 20 elements
atomic_numbers = list(range(1, 21))
symbols = ['H', 'He', 'Li', 'Be', 'B', 'C', 'N', 'O', 'F', 'Ne', 
           'Na', 'Mg', 'Al', 'Si', 'P', 'S', 'Cl', 'Ar', 'K', 'Ca']
# First ionization energies in kJ/mol
ionization_energies = [1312.0, 2372.3, 520.2, 899.5, 800.6, 1086.5, 1402.3, 1313.9, 1681.0, 2080.7,
                       495.8, 737.7, 577.5, 786.5, 1011.8, 999.6, 1251.2, 1520.6, 418.8, 589.8]

plt.figure(figsize=(12, 7))
plt.rcParams["font.family"] = "Bozon"

# Plotting the data
plt.plot(atomic_numbers, ionization_energies, marker='o', linestyle='-', color='salmon', markersize=6)

# Adding annotations
for i, symbol in enumerate(symbols):
    plt.annotate(symbol, 
                 (atomic_numbers[i], ionization_energies[i]), 
                 textcoords="offset points", 
                 xytext = (0,-20) if i+1 in [1,3,5,8,11,13,14,16,19] else (0,10), 
                 ha='center', 
                 fontsize=11,
                 fontweight='bold')

# Formatting the plot
plt.title('First Ionization Energy of the First 20 Elements', fontsize=16)
plt.xlabel('Atomic Number', fontsize=14)
plt.ylabel('First Ionization Energy (kJ/mol)', fontsize=14)
plt.xticks(atomic_numbers)
plt.grid(True, axis="y", ls="dashed")
plt.yticks([500,1000,1500,2000,2500])
plt.tight_layout()

# Save the plot
plt.savefig('ionization_energies.png', dpi=300)
```
