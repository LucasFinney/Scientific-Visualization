- Want to create an animation illustrating the idea of vector addititon for velocities
- One velocity going horizontally ("wind east") and one velocity going vertically ("plane's motion north")

```python
%%manim -qh -v WARNING Test
class Test(Scene):
    def construct(self):
        self.camera.background_color = WHITE
        # Create a plane and show its motion, wind, and net velocity
        # Make all the lines and labels in black color, except the net velocity which should be in red
        plane = SVGMobject(file_name="plane.svg").shift(DOWN*2)
        motion_line = Line(plane.get_center(), plane.get_center() + UP*4).add_tip().set_color(BLACK)
        motion_label = Text("Motion",font="Open Sans").next_to(motion_line, UP).set_color(BLACK)
        wind_line = Line(plane.get_center(), plane.get_center() + RIGHT*2).add_tip().set_color(BLACK)
        wind_label = Text("Wind",font="Open Sans").next_to(wind_line, RIGHT).set_color(BLACK)
        net_velocity = Line(plane.get_center(), plane.get_center() + UP*4 + RIGHT*2).add_tip().set_color(RED)
        net_label = Text("Net Velocity",font="Open Sans").next_to(net_velocity, UP).set_color(RED)
        # Show the motion of the plane and the effect of the wind on its trajectory. Motion without wind first, then add the wind and show the net velocity.
        self.play(FadeIn(plane))
        self.play(Create(motion_line),Write(motion_label))
        self.add(plane) # Readd the plane to the scene to ensure it's on top of the motion line
        self.play(MoveAlongPath(plane, motion_line), run_time=2)
        # Reposition the plane back to the start
        self.play(plane.animate.move_to(DOWN*2))
        # Move the motion lable to avoid overlap with the wind label
        self.play(motion_label.animate.next_to(motion_line, LEFT*1.5))
        self.play(Create(wind_line),Write(wind_label))
        self.play(Create(net_velocity),Write(net_label))
        self.add(plane) # Readd the plane to the scene to ensure it's on top of the motion line
        self.play(MoveAlongPath(plane, net_velocity), run_time=2)
        self.wait(2)
        self.play(FadeOut(plane), FadeOut(motion_line), FadeOut(motion_label), FadeOut(wind_line), FadeOut(wind_label), FadeOut(net_velocity), FadeOut(net_label))
```
